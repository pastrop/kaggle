```python
from openai import OpenAI
```

```python
from google.colab import userdata
fubu=userdata.get('fubu')
```

```python
input = "Create a portfolio maximising total annual return for a US retail investor over a 5-year horizon while ensuring the portfolio loses no more than 5% of invested capital in any single calendar year. The portfolio must satisfy a hard growth-asset ceiling of 75% and maintain a minimum 25% allocation to high-grade defensive assets at all times. Market regime assumptions baked in as of 14 June 2026: Fed funds target rate 3.50–3.75% (EFFR ~3.62%); futures markets price a gradual drift toward 3.8% by end-2026 with a ~30% probability of a hike by Q1-2027; 10-year Treasury yield ~4.47–4.50%; 2-year Treasury yield ~4.13%; 30-year Treasury yield ~4.97%; upward-sloping yield curve (10Y–2Y spread +40 bp, 10Y–3M spread +77 bp); S&P 500 at ~7,431 (off its early-June ATH of ~7,610, up ~10% YTD), elevated valuations at historically rare levels; VIX ~17.7 (moderating from a mid-March spike above 30 driven by Middle East conflict and energy-price surge; 52-week range 13.4–35.3); CPI and PPI at multi-year highs, inflation elevated and sticky; geopolitical risk elevated (active Middle East conflict, oil-price uncertainty, potential Iran deal). Given sticky inflation, the spec tilts defensively within the growth budget, favouring dividend/value equity, short-duration IG bonds, TIPS, and actively managed options overlays to cap downside."
```

```python
base_url = "https://api.sakana.ai/v1"
if not base_url.endswith("/v1"):
    base_url = f"{base_url}/v1"

client = OpenAI(
    api_key=fubu,
    base_url=base_url,
)

response = client.responses.create(
    model="fugu",
    tools=[{"type": "web_search"}],
    input=input,
)

print(response.output_text)
```
